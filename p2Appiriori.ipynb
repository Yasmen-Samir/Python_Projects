```python
import pandas as pd
from itertools import combinations
import itertools as it
```

```python
data0 = pd.read_excel('CoffeeShopTransactions.xlsx')
df = pd.DataFrame(data0)

df1 = df[df.columns[2]]
df2 = df.iloc[:,-3:]
df3 = df.iloc[:,-3:]

for i in range(0,3):
  df3[df3.columns[i]] = df2[df2.columns[i]].str.lower()
  df2[df2.columns[i]] = df3[df3.columns[i]].str.replace(" ","")

list1 = df1.values.tolist()
list2 = df2.values.tolist()

data = []

for d in list1:
  data.append([d] + [list2[d-1]])

print(data)
```

```python
# Step 2: Define a function to generate frequent itemsets based on the Apriori algorithm
def generate_frequent_itemsets(data, min_support):
  
    min_support_count = min_support * len(data)
    item_counts = {}
    for transaction in data:
        for item in transaction[1]:
            if item not in item_counts:
                item_counts[item] = 0
            item_counts[item] += 1

    frequent_items = {frozenset([item]): support for item, support in item_counts.items() if support >= min_support_count}
    current_items = frequent_items.keys()
    
    united_items = set()
    temp = list(frequent_items)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
          if(len(temp[i].union(temp[j]))) == 2:
            united_items.add(temp[i].union(temp[j]))

    item_counts2 = {}
    for i in united_items:
        item_counts2[i] = 0
        for transaction in data:
            temp = set(transaction[1])
            if(i.issubset(temp)):
                item_counts2[i]+=1

    frequent_items2 = {item: support for item, support in item_counts2.items() if support >= min_support_count}
    current_items2 = frequent_items2.keys()

    united_items2 = set()
    temp = list(frequent_items2)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
          if(len(temp[i].union(temp[j]))) == 3:
            united_items2.add(temp[i].union(temp[j]))

    item_counts3 = {}
    for i in united_items2:
        item_counts3[i] = 0
        for transaction in data:
            temp = set(transaction[1])
            if(i.issubset(temp)):
                item_counts3[i]+=1

    frequent_items3 = {item: support for item, support in item_counts3.items() if support >= min_support_count}
    current_items3 = frequent_items3.keys()

    return frequent_items3
```

```python
# Step 3: Define a function to generate association rules from the frequent itemsets
def generate_association_rules(frequent_items, min_confidence):
  
  for frequent in list(frequent_items):
      combsOf2 = [frozenset(q) for q in combinations(frequent,len(frequent)-1)]
      support_of_all_rule = frequent_items[frequent]
                                           
      for antecedent in combsOf2:
        Consequent = frequent-antecedent
        support_of_antecedent = 0
        support_of_Consequent = 0

        for transaction in data:
            temp = set(transaction[1])
            if(antecedent.issubset(temp)):
                support_of_antecedent+=1
            if(Consequent.issubset(temp)):
                support_of_Consequent+=1

        confedence = support_of_all_rule/support_of_antecedent
        confedence2 = support_of_all_rule/support_of_Consequent

        if(confedence > min_confidence):
          print(str(list(antecedent))+" -> "+str(list(Consequent))+" = "+str(confedence))
        if(confedence2 > min_confidence):
          print(str(list(Consequent))+" -> "+str(list(antecedent))+" = "+str(confedence2))
```

```python
# Step 4: Ask the user for the minimum support and confidence values
min_support = float(input('Enter the minimum support : '))
min_confidence = float(input('Enter the minimum confidence: '))

# Step 5: Use the above functions to generate frequent itemsets and association rules based on the user input and print them

print('\nFrequent Itemsets with their support:\n')

frequent_items = generate_frequent_itemsets(data, min_support)
for itemset, support in frequent_items.items():
    print(f'{set(itemset)}: {support/len(data)}')

print('\nAssociation Rules with confidence:\n')

association_rules = generate_association_rules(frequent_items, min_confidence)
```
