```python
import pandas as pd
import numpy as np
import random
```

```python
data0 = pd.read_csv('crime_data.csv')
df = pd.DataFrame(data0)

df1 = df[df.columns[0]]
df2 = df.iloc[:,-4:]

list1 = df1.values.tolist()
list2 = df2.values.tolist()

data = {}

for d in range(len(list2)):
  data[tuple(list2[d])] = list1[d]

print(data)
```

```python
def Distance2Point(point1, point2):
  return sum(abs(a - b) for a, b in zip(point1, point2))
```

```python
def KMean(data, num_of_clusters):

  max_iteration = 2
  centroids = {}
  transactions = []

  for d in data:
    transactions.append(list(d))

  for i in range(num_of_clusters): 
    centroids[i] = random.choice(transactions)            #c0 = the first list in data

  classes = {}
  classes2 = {}                        #initialize clusters with empty list
    
  for iteration in range(max_iteration):   # 0-----2
      classes = {}
      classes2 = {}
      outliers = {}

      for classKey in range(num_of_clusters): # 0------4
            classes[classKey] = []
            classes2[classKey] = []
            outliers[classKey] = []

      for dataPoint in transactions: 
            Distance = []
            for centroid in centroids:    #  c0 = the first list in data  
              dis = Distance2Point(dataPoint, centroids[centroid])
              Distance.append(dis)   # distance between the first datapoint and every centroid

            minDis = min(Distance)
            minDisIndex = Distance.index(minDis)    # index  ------->  0,1,2,3
            classes[minDisIndex].append(dataPoint)  # classes = {0:[[2.33,1.32,6.44,6],[2,5,1.88,6.2],[2,6,8,4]]}
            classes2[minDisIndex].append(data[tuple(dataPoint)])
            
            if(minDis > 70): 
              outliers[minDisIndex].append(data[tuple(dataPoint)])
                
      for classKey in classes:
            classData = classes[classKey]
            NewCentroid = np.mean(classData, axis = 0)
            centroids[classKey] = NewCentroid

  return centroids, classes2 , outliers
```

```python
num_of_clusters = int(input("please enter number of clusters \n"))

centroids, classes, outliers = KMean(data, num_of_clusters)

print("classes : \n", classes)
print()
print("centroids : \n ", centroids)
print()
print("outliers : \n ", outliers)
```
